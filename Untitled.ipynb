```python
import pandas as pd
import numpy as np
from Pfeature.pfeature import aac_wp
from sklearn.ensemble import RandomForestClassifier, RandomForestClassifier

def to_df(path, out):
    aac_wp(path, out)
    return pd.read_csv(out)
```

```python
train_ne_df = to_df('train_set/ne_cd_hit.txt', 'train_set/train_ne.csv')
train_po_df = to_df('train_set/po_cd_hit.txt', 'train_set/train_po.csv')
test_ne_df = to_df('test_set/ne_cd_hit.txt', 'test_set/test_ne.csv')
test_po_df = to_df('test_set/po_cd_hit.txt', 'test_set/test_po.csv')
```

```python
train_ne_df['id'] = 0
train_po_df['id'] = 1
test_ne_df['id'] = 0
test_po_df['id'] = 1
```

```python
train_df = pd.concat([train_ne_df, train_po_df], ignore_index=True)
train_df = train_df.sample(frac=1).reset_index(drop=True)
train_df
```

```python
test_df = pd.concat([test_ne_df, test_po_df], ignore_index=True)
test_df = test_df.sample(frac=1).reset_index(drop=True)
test_df
```

```python
X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]
```

```python
X_test = test_df.iloc[:, :-1]
y_test = test_df.iloc[:, -1]
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42, criterion='gini', max_features='sqrt')
model.fit(X_train, y_train)
```

```python
def accuracy_score(y, y_pred):
    res = np.logical_not(np.logical_xor(y, y_pred))
    acc = np.count_nonzero(res == 1) / len(y)
    return acc
    
y_pred = model.predict(X_test)
y_pred
accuracy = accuracy_score(y_test, y_pred)
f"Accuracy: {accuracy}"
```
